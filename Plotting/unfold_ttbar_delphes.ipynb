```python
def find_pattern(pattern, matrix): 
    
    solns_matrix = [[] for n in trange(len(matrix))]
    
    N = len(pattern)
    for i in trange(len(matrix)):
        values = matrix[i]
        possibles = np.where(values == pattern[0])[0]
        
        solns = []
        for p in possibles:
            check = values[p:p+N]
            if np.all(check == pattern):
                solns.append(p)
        
        solns_matrix[i] = solns   
    return solns_matrix
```
